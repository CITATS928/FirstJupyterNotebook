```python
most_common_city = chosen_file['City (Administrative Location)'].value_counts().idxmax()
count_most_common_city = chosen_file['City (Administrative Location)'].value_counts().max()
print(f"The city with the most museums is {most_common_city} with {count_most_common_city} museums.")
```

```python
least_common_city = chosen_file['City (Administrative Location)'].value_counts().idxmin()
count_least_common_city = chosen_file['City (Administrative Location)'].value_counts().min()
print(f"The city with the least museums is {least_common_city} with {count_least_common_city} museums.")
```

```python
city_museum_counts = chosen_file['City (Administrative Location)'].value_counts()
print(city_museum_counts)
```

```python
zoo_aquarium_count=chosen_file['Museum Type'].str.contains('zoo|aquarium',case=False,na=False).sum()
print(zoo_aquarium_count)
zoo_count=chosen_file['Museum Name'].str.contains('zoo',case=False,na=False).sum()
print(zoo_count)
aquariums_count=chosen_file['Museum Name'].str.contains('aquarium',case=False,na=False).sum()
print(aquariums_count)
```

```python
highest_revenue_index = chosen_file['Revenue'].idxmax()
highest_revenue_museum = chosen_file.loc[highest_revenue_index, 'Museum Name']
highest_revenue_amount = chosen_file.loc[highest_revenue_index, 'Revenue']

print(f"The museum with the highest revenue is {highest_revenue_museum} with a revenue of {highest_revenue_amount}.")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
art_museums = chosen_file[chosen_file['Museum Type'].str.contains('art', case=False, na=False)]
```

```python
usa = gpd.read_file("ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp")
states = usa[usa['CONTINENT'] == 'North America']
```

```python
art_museums_gdf = gpd.GeoDataFrame(
    art_museums,
    geometry=gpd.points_from_xy(art_museums['Longitude'], art_museums['Latitude']),
    crs="EPSG:4326"
)

fig, ax = plt.subplots(1, 1, figsize=(15, 10))
states.plot(ax=ax, color='lightgrey', edgecolor='black')
art_museums_gdf.plot(ax=ax, color='blue', markersize=10, alpha=0.6, label='Art Museums')

plt.title('Distribution of Art Museums in the USA')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```
